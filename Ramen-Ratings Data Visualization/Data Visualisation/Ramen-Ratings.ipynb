 ### Top 10 Ramen 

Let's have a look at Top 10 Ramens

```python
# Making a copy of original dataframe 
top10 = dataframe.copy()

# Dropping the data which have blank value in 'Top Ten' Column. 
# We will only analyse the Ramen which has come in Top Ten list
top10 = top10.dropna(subset=['Top Ten']).copy()

top10
```

### We can see that there are some anomalies in the dataset. We found '\n' or newline in our data. Let's filter it.

```python
#Removing the Data which have a newline in 'Top Ten' column
top10 = top10[top10['Top Ten'] != '\n']

#Sorting the data by year
top10 = top10.sort_values('Top Ten' ) 

top10
```

### Which brands have come in Top 10 most number of times ?
### Let's find it.

```python
# Using value_counts()
top10['Brand'].value_counts()
```

### Prima and Prima Taste refer to the same brand - https://www.primataste.com/

### Let us replace all the 'Prima' by 'Prima Test' in our 'top10' dataframe.

```python
#Replacing 'Prima' by 'Prima Taste'
top10['Brand']= top10['Brand'].replace('Prima','Prima Taste')
```

```python
top10['Brand'].value_counts()
```

### Prima Taste brand has come in Top Ten List most number of times

 ### Top Brands by rating

```python
# Making a copy of original dataframe 
top_rating = dataframe.copy()

#Replacing 'Prima' by 'Prima Taste'
top_rating['Brand']= top_rating['Brand'].replace('Prima','Prima Taste')
```

```python
#Sorting the rows in descending order of Star Ratings
top_rating = top_rating.sort_values('Stars',ascending = False)
top_rating
```

### Let's remove the Unrated Data

```python
# Ignoring the data which is unrated
top_rating = top_rating[top_rating['Stars']!='Unrated']
top_rating
```

### Let's look at top 50 Rated brands

### We will compare the average ratings of all the brands.

```python
brands = list(top_rating['Brand'].unique())
counter = [0.0]*355

#create dictionary to count all ratings and then save the averages
brands_cnt = dict(zip(brands, counter)) 

for brand in brands:
    brands_data = top_rating[top_rating['Brand'] == brand]
    for star in brands_data['Stars']:
        #count all ratings
        brands_cnt[brand] += float(star)
    #average    
    brands_cnt[brand] /= len(brands_data) 
```

```python
#list for saving the brand name and its average rating
top50ratings = [] 
for key, values in brands_cnt.items():
    top50ratings.append([key,values])

#print the top 50 ramen ratings by brand
top50ratings = sorted(top50ratings, key = lambda x : x[1], reverse = True) #sorting values in descending order
top50ratings
for i in range(50):
    print('#{:<3}{:25} {}'.format(i+1, top50ratings[i][0], round(top50ratings[i][1],2)))
```

## Saving Data using Pickle

```python
file1 = "Top 50 Ratings.pkl"
with open(file1, 'wb') as file:  
    pickle.dump(top50ratings, file)

file2 = "Top 10 Brands.pkl"

with open(file2, 'wb') as file:  
    pickle.dump(top10['Brand'].value_counts(), file)
```
