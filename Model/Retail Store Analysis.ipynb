# Data Visualization

```python
eda.plot.hist(figsize=(10, 30), subplots=True); # To gain familiarity with the data
```

```python
sns.pairplot(eda);# Visualizing the data as a whole using the pair plot
```

Most used Ship Mode

```python
mode = eda['Ship Mode'].value_counts()
mode
```

```python
mode.plot(x = 'Ship Mode', kind = 'bar',color = ['blue','red','orange','green'])
plt.title('Ship Mode');
```

Major Segment for which orders are shipped

```python
segment = eda['Segment'].value_counts()
segment
```

```python
segment.plot(kind = 'pie');
plt.title('Segment')
plt.ylabel(' ');
```

Top 10 Cities with most orders shipped

```python
city = eda['City'].value_counts().nlargest(n=10)
city
```

```python
city.plot(x = 'City', kind = 'bar', color = 'orange')
plt.title('City');
```

Top 10 States with most orders shipped

```python
state = eda['State'].value_counts().nlargest(n=10)
state
```

```python
state.plot(x = 'State', kind = 'bar', color = 'red')
plt.title('State');
```

Region with most shippment

```python
region = eda['Region'].value_counts()
region
```

```python
sns.countplot(eda['Region']);
```

Most shipped Category

```python
category = eda['Category'].value_counts()
category
```

```python
category.plot(kind = 'pie');
plt.title('Category')
plt.ylabel(' ');
```

Most shipped Sub-Category

```python
subCategory = eda['Sub-Category'].value_counts()
subCategory
```

```python
plt.figure(figsize=(20,10))
sns.countplot(eda['Sub-Category']);
```

Sales and Profit made across various Ship Mode

```python
eda.groupby("Ship Mode")[[ 'Sales','Profit']].sum().plot.bar(color = ["green","red"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

Sales and Profit made across various Segment

```python
eda.groupby("Segment")[[ 'Sales','Profit']].sum().plot.bar(color = ["blue","purple"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show();
```

Top 10 Cities with maximum sales

```python
city_sales= eda.groupby("City").Sales.sum().nlargest(n=10)
city_sales
```

```python
plt.figure(figsize=(10,10))
city_sales.plot(x = 'City', kind = 'bar', color = 'Black')
plt.title('Max sales');
plt.ylabel("Sales");
```

Top 10 Cities with maximum profit

```python
city_profit= eda.groupby("City").Profit.sum().nlargest(n=10)
city_profit
```

```python
plt.figure(figsize=(10,10))
city_profit.plot(x = 'City', kind = 'bar', color = 'yellow')
plt.title('Max Profit');
plt.ylabel("Profit");
```

Top 10 States with maximum sales

```python
state_sales= eda.groupby("State").Sales.sum().nlargest(n=10)
state_sales
```

```python
plt.figure(figsize=(10,10))
state_sales.plot(x = 'State', kind = 'bar', color = 'orange')
plt.title('Max sales');
plt.ylabel("Sales");
```

Top 10 States with maximum profit

```python
state_profit = eda.groupby("State").Profit.sum().nlargest(n=10)
state_profit
```

```python
plt.figure(figsize=(10,10))
state_profit.plot(x = 'State', kind = 'bar', color = 'purple')
plt.title('Max Profit');
plt.ylabel("Profit");
```

Sales and Profit across various regions

```python
region_sales_profit= eda.groupby("Region")[[ 'Sales','Profit']].sum()
region_sales_profit
```

```python
region_sales_profit.plot.bar(color = ["blue","red"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

Sales and Profit across various Categories

```python
category_sales_profit= eda.groupby("Category")[[ 'Sales','Profit']].sum()
category_sales_profit
```

```python
category_sales_profit.plot.bar(color = ["green","orange"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

Sales and Profit across various Categories

```python
Sub_Category_sales_profit= eda.groupby("Sub-Category")[[ 'Sales','Profit']].sum()
Sub_Category_sales_profit
```

```python
Sub_Category_sales_profit.plot.bar(color = ["Blue","Purple"],figsize= (20,10))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

Discount Across Various Categories

```python
category_discount = eda.groupby("Category").Discount.sum()
category_discount
```

```python
category_discount.plot(kind = 'pie');
plt.title('Category')
plt.ylabel(' ');
```

Discount Across Various Sub_Category

```python
Sub_Category_discount = eda.groupby("Sub-Category").Discount.sum()
Sub_Category_discount
```

```python
plt.figure(figsize=(10,10))
Sub_Category_discount.plot(x = 'Sub-Category', kind = 'bar', color = 'purple')
plt.title('Discount Across Various Sub_Category');
plt.ylabel("Discount");
```

3. More focus should be put on selling Office Supplies and Technology as they lead to profit.
4. Less focus should be put on selling Furnitures as they account for losses.
5. New york and Los Angeles has highest profit.
6. We can conclude that people residing in the west region order more as compared to other regions,so in order to get more profit in this area we need to decrease the Discount on Furniture Category.
