```python
#import libraries
%pylab inline
import pandas
import seaborn
```

```python
data = pd.read_csv('uber-raw-data.csv')
```

```python
data.head()
```

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
data.head()
```

```python
def getdom(df):
    return df.day

data['dom'] = data['Date/Time'].map(getdom)
```

```python
data.tail()
```

```python
def getweekday(df):   
    return df.weekday()

data['weekday'] = data['Date/Time'].map(getweekday)

def gethour(df):
    return df.hour

data['hour'] = data['Date/Time'].map(gethour)

data.tail()
```

# analyze the DoM

```python
hist(data.dom, bins=30, rwidth=.8, range=(0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - uber')
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
bar(range(1, 31), by_date)
```

```python

by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python

bar(range(1, 31), by_date_sorted)
xticks(range(1,31), by_date_sorted.index)
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - uber')
;
```

# Analyze the hour

```python
hist(data.hour, bins=24, range=(.5, 24))
```

# Analyze the weekday

```python
hist(data.weekday, bins=7, range =(-.5,6.5), rwidth=.8, color='#AA6666', alpha=.4)
xticks(range(7), 'Mon Tue Wed Thu Fri Sat Sun'.split())
```

# Heatmap

```python
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()

seaborn.heatmap(by_cross)
```
