```python
#import libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel("Online_Retail.xlsx")
df.head()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df = df[-df['InvoiceNo'].str.contains('C')]
df
```

```python
basket = (df[df['Country'] =="France"]
         .groupby(['InvoiceNo', 'Description'])['Quantity']
         .sum().unstack().reset_index().fillna(0)
         .set_index('InvoiceNo'))
basket
```

```python
def encode(a):
    if a<= 0:
        return 0
    if a>= 1:
        return 1
    
basketset = basket.applymap(encode)
basketset.drop('POSTAGE', inplace=True, axis=1)
basketset
```

```python
frequentitem = apriori(basketset, min_support=0.07, use_colnames=True)
rules = association_rules(frequentitem, metric="lift",min_threshold=1)
rules.head()
```

```python
rules[(rules['lift'] >= 6) & (rules['confidence'] >= 0.8)]
```
