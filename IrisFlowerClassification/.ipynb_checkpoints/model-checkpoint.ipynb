# Gathering training and testing data

```python
X = df.drop(['Id', 'Species'], axis=1)
y = df['Species']
```

```python
X_features = X
X = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Check for different values of neighbours

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
k_range = list(range(1, 31))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
# Checking for accuracy

plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

```python
# Finalised Model

knn = KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# Exporting Model
joblib.dump(knn, "./exported-model/Knn-classifier-model.joblib")
```

```python
# Accuracy using exported model
model=joblib.load('./exported-model/Knn-classifier-model.joblib')
predict = model.predict(X_test)
print(metrics.accuracy_score(y_test, predict))
```
